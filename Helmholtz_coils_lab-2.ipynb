```julia
using Statistics
```

```julia
using LsqFit
```

```julia
model(x, p) = p[1]*x
```

```julia
xdata = vcat(x1,x2,x3,x4,x5,x6)
ydata = vcat(y1, y2, y3, y4, y5, y6)
```

```julia
fit = curve_fit(model, [e.val for e in xdata.*1e12], ydata, [0.17])
```

```julia
bestfit(x) = coef(fit)[1]*x*1e12
```

```julia
m = coef(fit)[1]*1e12 ± stderror(fit)[1]*2e12 #2σ uncertainty
```

```julia
expected(x) = (1.758820e11)*x
```

```julia
p1 = scatter(x1,y1, ylim=(0, 150), label="First test")
p1 = scatter!(x2,y2, ylim=(0, 150), label="Second test")
p1 = scatter!(x3,y3, ylim=(0, 150), label="Third test")
p1 = scatter!(x4,y4, ylim=(0, 150), label="Fourth test")
p1 = scatter!(x5,y5, ylim=(0, 150), label="Fifth test")
p1 = scatter!(x6,y6, ylim=(0, 150), label="Sixth test")
p1 = plot!([0.0, 1.0e-9], bestfit.([0.0, 1.0e-9]), label="Best fit line")
p1 = plot!([0.0, 1.0e-9], expected.([0.0, 1.0e-9]), label="Expected line")

p1 = xlims!(0, 1.0e-9)

p1 = annotate!(
  6.0e-10,
  30,
  "Best fit: "*L"y = ("*string(round(m, digits=-9))*L"\ \dfrac{C}{kg})x"
)

SStot = sum((ydata.-mean(ydata)).^2)
SSres = sum((ydata.-bestfit.(xdata)).^2)
Rsq = 1-SSres/SStot

plt = annotate!(
  6.0e-10,
  20,
  latexstring("\$R^2 = \$"*string(round(Rsq.val, digits=4)))
)

p1 = title!("\$2V\$ vs \$B^2r^2\$")
p1 = ylabel!("\$2V\$ (\$V\$)")
p1 = xlabel!("\$B^2r^2\$ (\$T^2m^2\$)")

display(p1)
```
