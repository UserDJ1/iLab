```julia
using Plots, FileIO, ImageMagick, Cairo, LinearAlgebra, Distributions, Random, SpecialFunctions, Parameters, ProgressMeter, Dates, LinkedLists, LsqFit,
    Measurements, LaTeXStrings;
gr();
```

```julia
xdata = [130 ± 0.1,125 ± 0.1,120 ± 0.1,115 ± 0.1,110 ± 0.1,105 ± 0.1,100 ± 0.1,95 ± 0.1,90 ± 0.1,85 ± 0.1,80 ± 0.1,75 ± 0.1,70 ± 0.1,65 ± 0.1,60 ± 0.1,55 ± 0.1,50 ± 0.1,45 ± 0.1,40 ± 0.1,35 ± 0.1,30 ± 0.1,25 ± 0.1,20 ± 0.1,18 ± 0.1,16 ± 0.1,14 ± 0.1,12 ± 0.1,10 ± 0.1,8 ± 0.1,6 ± 0.1]
x = [e.val for e in xdata]

ydata = [0.325 ± 0.0005,0.325 ± 0.0005,0.325 ± 0.0005,0.345 ± 0.0005,0.333 ± 0.0005,0.357 ± 0.0005,0.317 ± 0.0005,0.341 ± 0.0005,0.341 ± 0.0005,0.337 ± 0.0005,0.321 ± 0.0005,0.31 ± 0.0005,0.333 ± 0.0005,0.31 ± 0.0005,0.321 ± 0.0005,
0.313 ± 0.0005,0.29 ± 0.0005,0.286 ± 0.0005,0.274 ± 0.0005,0.247 ± 0.0005,0.212 ± 0.0005,0.149 ± 0.0005,0.012 ± 0.0005,-0.089 ± 0.0005,-0.234 ± 0.0005,-0.453 ± 0.0005,-0.821 ± 0.0005,-1.446 ± 0.0005,-2.624 ± 0.0005,-5.154 ± 0.0005]
y = [e.val for e in ydata];
```

```julia
@. model(x, p) = p[1]/(x + p[2])^3 + 0.047
fit = curve_fit(model, [e.val for e in xdata], [e.val for e in ydata], [-10000, 0.1]);
```

```julia
m = coef(fit)
```

```julia
σ = stderror(fit)
```

```julia
bestfit(x) = m[1]/(x + m[2])^3 + 0.047
```

```julia
SStot = sum((y.-mean(y)).^2)
SSres = sum((y.-bestfit.(x)).^2)
Rsq = 1-SSres/SStot

plot(xdata, ydata, label="Experimental", seriesalpha=0.8)
plot!([e.val for e in xdata], bestfit.([e.val for e in xdata]), label="Theoretical", linestyle=:dash, color="red", seriesalpha=0.8)
annotate!(
  75,
  -0.2,
  latexstring("\$R^2 = \$"*string(round(Rsq, digits=4)))
)
ylabel!("Magnetic Field (\$mT\$)")
xlabel!("Distance (\$m\$)")
title!("Magnetic Field Strength vs Distance")
```
