```julia
using Plots
```

```julia
const μ₀ = 1.25663706212e-6 #N/A^2
```

```julia
B(ζ, I, N, R) = ( ( μ₀*I*N)/(2*R))*((1/(ζ^2 - ζ + 5/4)^(3/2) ) + (1/(ζ^2 + ζ + 5/4)^(3/2) ))
```

```julia
I = 10 #A
R = 0.030 #m
N = 72 #turns
z = range(start=-R, stop=R, length=1000) #m
ζ = z./R #unitless
```

```julia
plt = plot(z, B.(ζ, I, N, R), label="B(ζ)")
```
