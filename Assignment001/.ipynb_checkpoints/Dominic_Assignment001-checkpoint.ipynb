### <font color='green'>Q4 Answer</font>

```julia
using Measurements

r = 2.0 ± 0.1

v = (4/3)*π*r^3
```

Volume of the sphere $v = \frac{4}{3}\pi r^3 = 33.5 \pm 5.0 \ m^3$\
\
Volume uncertainty $\Delta v = v\sqrt{\left(\frac{3\Delta r}{r}\right)^2} = (33.5m^3)\left|\frac{(3)(0.1m)}{2.0m}\right| = 5.0m^3$

## <font color='purple'>Question 6</font>
If a stone is thrown vertically upward with speed $v$, it should rise to a height $h$ given by $v^2 = 2gh$. In particular, $v^2$ should be proportional to $h$. To test this proportionality, a student on Planet X measures $v^2$ and $h$ for seven different throws and obtains the results shown here:

| $h$             | $v^2$               |
|-----------------|---------------------|
| $\left (\,\pm\, 0.05\,\mathrm{m}\right )$  | $\left (\,\pm 3\,\mathrm{m}^2/\mathrm{s}^2\right )$     |
| $0.40$          | $7$                 |
| $0.80$          | $17$                |
| $1.4$           | $25$                |
| $2.0$           | $38$                |
| $2.6$           | $45$                |
| $3.4$           | $62$                |
| $3.8$           | $72$                |

a. Make a plot of $v^2$ vs $h$, including vertical and horizontal error bars using either As usual, label your axes, and scale the axes suitably. Is your plot consistent with the prediction that $v^2 \propto h$?<br>

b. The slope of your graph should be $2g$. To find the slope, draw the best fit straight line by performing a linear fit to the data. 
Use the [LsqFit.jl](https://github.com/JuliaNLSolvers/LsqFit.jl) package to do this. What is your experimental value of the slope and it's $1\sigma$ uncertainty?

### <font color='green'>Q6 Answers</font>



a.

```julia
h = [0.40 ± 0.05, 0.80 ± 0.05, 1.4 ± 0.05, 2.0 ± 0.05, 2.6 ± 0.05, 3.4 ± 0.05, 3.8 ± 0.05];
vsquared = [7 ± 3, 17 ± 3, 25 ± 3, 38 ± 3, 45 ± 3, 62 ± 3, 72 ± 3];
```

###  <font color='red'> In the cell below, do not connect points by discontinuous sloped segments. Why?</font>

```julia
plot(h, vsquared)
title!("\$v^2\$ vs \$h\$")
xlabel!("\$h\$ (\$m\$)")
ylabel!("\$v^2\$ (\$m^2/s^2\$)")
```

This plot appears consistent with the linear relationship between $v^2$ and $h$.\
\
b.

```julia
using LsqFit

#@. model(x, p) = p[1]*x
model(x, p) = p[1]*x

fit = curve_fit(model, [e.val for e in h], [e.val for e in vsquared], [19.6])
```

```julia
m= coef(fit)[1]
```

```julia
line(x) = m*x
```

###  <font color='red'> In the cell below, do not connect points by discontinuous sloped segments. </font>

```julia
plot(h, vsquared, label="Experimental")
plot!([e.val for e in h], line.([e.val for e in h]), label="Best fit line", seriesalpha=0.5, color="red", linestyle=:dash)
title!("\$v^2\$ vs \$h\$ experimental data with best fit line")
xlabel!("\$h\$ (\$m\$)")
ylabel!("\$v^2\$ (\$m^2/s^2\$)")
```

```julia
sigma = stderror(fit)[1]
```

The best fit line is $v' \approx 18.4h$.\
The $1\sigma$ error on the slope is $0.3$.
